```python
sns.histplot(df["age"])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
```

```python
sns.histplot(df["bmi"])
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Frequency")
```

```python
sns.histplot(df["children"])
plt.title("Children Distribution")
plt.xlabel("Children")
plt.ylabel("Frequency")
```

```python
sns.histplot(df["expenses"])
plt.title("expenses Distribution")
plt.xlabel("expenses")
plt.ylabel("Frequency")
```

```python
sns.countplot(df["sex"])
plt.title("Gender count plot")
plt.xlabel("Frequency")
plt.ylabel("Gender")
```

```python
sns.countplot(df["smoker"])
plt.title("Smoker count")
plt.xlabel("Smoker")
plt.ylabel("Count")
```

```python
sns.countplot(df["region"])
plt.title("Region count plot")
plt.xlabel("Frequency")
plt.ylabel("Region")
```

```python
sns.heatmap(df.corr(numeric_only= True), annot= True)
plt.title("Correlation Heatmap")
```

```python
print("Average insurance Expenses:")
print(df['expenses'].mean())

print("\nMaximum Expenses:")
print(df['expenses'].max())

print("\nMinimum Expenses:")
print(df['expenses'].min())

print("\nAverage Expenses for smokers vs Non-smokers:")
print(df.groupby('smoker')['expenses'].mean())
```

```python
sns.boxplot(x = df['smoker'], y= df['expenses'])
plt.title("Insurance charges by smoker")
plt.xlabel("Smoker")
plt.ylabel("Charges")
```

```python
sns.boxplot(x = df['sex'], y= df['expenses'])
plt.title("Insurance charges by sex")
plt.xlabel("Sex")
plt.ylabel("Charges")
```

```python
sns.boxplot (x = df["sex"], y= df["expenses"], hue = df["smoker"])
plt.title("Insurance charges by sex and smoker")
plt.xlabel("Sex")
plt.ylabel("Charges")
```

```python
# Analysis Summary:-

# 1. Average Age and BMI
# - The average age of the customers was calculated using the mean() function.
# - The average BMI was also calculated using the mean() function.
# - The average age of the customers is approximately 39 years.
# - The average BMI of the customers is approximately 30.
# - These values represent the overall average age and body mass index of all customers.

# 2. Effect of Smoking on Insurance Charges
# - The dataset was grouped using groupby("smoker"), and the average insurance charges were calculated using the mean() function.
# - The analysis shows that smokers pay significantly higher insurance charges than non-smokers.
# - Smoking has the strongest impact on insurance charges among all the variables analyzed.

# 3. Region with the Highest Number of Customers
# - The Southeast region has the highest number of customers.
# - The value_counts() function was used on the region column to count the number of customers in each region.

# 4. Observations from the Visualizations
# - Distribution plots show the spread of age, BMI, children, and insurance charges.
# - A heatmap was created using corr() and sns.heatmap() to identify correlations between numerical variables.
# - The boxplot shows that smokers generally have much higher insurance charges than non-smokers.
# - Overall, smoking has the greatest impact on insurance charges, while age and BMI also influence insurance charges to some extent.
```
